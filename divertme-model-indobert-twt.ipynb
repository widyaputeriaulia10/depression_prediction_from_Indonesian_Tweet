```python
import pandas as pd
import xml.etree.ElementTree as ET
import xml.etree.cElementTree as et
from lxml import etree as etree
```

```python
def extract_from_xml(path):
    tree = et.parse(path)
    root = tree.getroot()
    child = ET.Element('child')
    root.append(child)
    return(root)
```

```python
def xml_to_dataframe(root):
    cont_id = []
    content = []
    label = []
    score_prd = []
    hashtagging = []
    topic_list = []
    create_date_time = []
    topic_feature = []
    trending_rank = []
    coverage_list = []
    date_trend = []
    time_trend = []
    for content_id in root.iter('content_id'):
        cont_id.append(content_id.text)
    for cont in root.iter('content'):
        content.append(cont.text)
    for lb in root.iter('prediction'):
        label.append(lb.text)
    for scr_pred in root.iter('score_pred'):
        score_prd.append(scr_pred.text)
    for tags in root.iter('hashtags'):
        hashtagging.append(tags.text)
    for top in root.iter('topics'):
        topic_list.append(top.text)
    for top_feats in root.iter('score'):
        topic_feature.append(top_feats.text)
    for date_time in root.iter('create_date_time'):
        create_date_time.append(date_time.text)
    for trd_rank in root.iter('trending_rank'):
        trending_rank.append(trd_rank.text)
    for trd_cover in root.iter('trending_coverage'):
        coverage_list.append(trd_cover.text)
    for trd_date in root.iter('trending_date'):
        date_trend.append(trd_date.text)
    for trd_time in root.iter('trending_time'):
        time_trend.append(trd_time.text)
    new_file = {"content_id":cont_id, "content":content,"hashtags":hashtagging, 'prediction':label, "score_pred":score_prd,"topics":topic_list,
               "score":topic_feature, "create_date_time":create_date_time, "trending_rank":trending_rank,
               "trending_coverage":coverage_list, "trending_date":date_trend, "trending_time":time_trend}
    new_ = pd.DataFrame(new_file)
    return(new_)
```

```python
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    
def count_param(module, trainable=False):
    if trainable:
        return sum(p.numel() for p in module.parameters() if p.requires_grad)
    else:
        return sum(p.numel() for p in module.parameters())
    
def get_lr(optimizer):
    for param_group in optimizer.param_groups:
        return param_group['lr']

def metrics_to_string(metric_dict):
    string_list = []
    for key, value in metric_dict.items():
        string_list.append('{}:{:.5f}'.format(key, value))
    return ' '.join(string_list)
```

```python
train_dataset_path = './train_ori.csv'
valid_dataset_path = './val_ori.csv'
test_dataset_path = './test_ori.csv'
```

## fungsi dataset loader dari utils IndoNLU


```python
train_dataset = DocumentSentimentDataset(train_dataset_path, tokenizer, lowercase=True)
valid_dataset = DocumentSentimentDataset(valid_dataset_path, tokenizer, lowercase=True)
test_dataset = DocumentSentimentDataset(test_dataset_path, tokenizer, lowercase=True)

train_loader = DocumentSentimentDataLoader(dataset=train_dataset, max_seq_len=128, batch_size=16, num_workers=2, shuffle=True)  
valid_loader = DocumentSentimentDataLoader(dataset=valid_dataset, max_seq_len=128, batch_size=16, num_workers=2, shuffle=False)  
test_loader = DocumentSentimentDataLoader(dataset=test_dataset, max_seq_len=128, batch_size=16, num_workers=2, shuffle=False)
```

```python
w2i, i2w = DocumentSentimentDataset.LABEL2INDEX, DocumentSentimentDataset.INDEX2LABEL
print(w2i) #word to index
print(i2w) #index to word
```

```python
# Tentukan optimizer
optimizer = optim.Adam(model.parameters(), lr=1.5e-6, eps = 1e-08)
model = model.cuda()
```

```python
n_epochs = 5
for epoch in range(n_epochs):
    model.train()
    torch.set_grad_enabled(True)
 
    total_train_loss = 0
    list_hyp_train, list_label = [], []

    train_pbar = tqdm(train_loader, leave=True, total=len(train_loader))
    for i, batch_data in enumerate(train_pbar):
        # Forward model
        loss, batch_hyp, batch_label = forward_sequence_classification(model, batch_data[:-1], i2w=i2w, device='cuda')

        # Update model
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        tr_loss = loss.item()
        total_train_loss = total_train_loss + tr_loss

        # Hitung skor train metrics
        list_hyp_train += batch_hyp
        list_label += batch_label

        train_pbar.set_description("(Epoch {}) TRAIN LOSS:{:.4f} LR:{:.8f}".format((epoch+1),
            total_train_loss/(i+1), get_lr(optimizer)))

    metrics = document_sentiment_metrics_fn(list_hyp_train, list_label)
    print("(Epoch {}) TRAIN LOSS:{:.4f} {} LR:{:.8f}".format((epoch+1),
        total_train_loss/(i+1), metrics_to_string(metrics), get_lr(optimizer)))

    # Evaluate di validation set
    model.eval()
    torch.set_grad_enabled(False)
    
    total_loss, total_correct, total_labels = 0, 0, 0
    list_hyp, list_label = [], []

    pbar = tqdm(valid_loader, leave=True, total=len(valid_loader))
    for i, batch_data in enumerate(pbar):
        batch_seq = batch_data[-1]        
        loss, batch_hyp, batch_label = forward_sequence_classification(model, batch_data[:-1], i2w=i2w, device='cuda')
        
        # Hitung total loss
        valid_loss = loss.item()
        total_loss = total_loss + valid_loss

        # Hitung skor evaluation metrics
        list_hyp += batch_hyp
        list_label += batch_label
        metrics = document_sentiment_metrics_fn(list_hyp, list_label)

        pbar.set_description("VALID LOSS:{:.4f} {}".format(total_loss/(i+1), metrics_to_string(metrics)))
        
    metrics = document_sentiment_metrics_fn(list_hyp, list_label)
    print("(Epoch {}) VALID LOSS:{:.4f} {}".format((epoch+1),
        total_loss/(i+1), metrics_to_string(metrics)))
```
